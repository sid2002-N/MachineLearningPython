```python
model_id1 = "stabilityai/stable-diffusion-3.5-large"

pipe = StableDiffusion3Pipeline.from_pretrained(model_id1, torch_dtype=torch.float16)
pipe = pipe.to("cpu")


```

```python
prompt ="A sleek, futuristic sports car parked on a glowing city street at night. The car has a streamlined, aerodynamic design with sharp angles and smooth curves, featuring a glossy metallic finish in a deep midnight blue. Its wheels are large, with glowing neon rims that reflect the light of the surrounding skyscrapers. The car's headlights are sharp, with a sharp, futuristic glow that cuts through the dark street. The body is accented with subtle glowing lines, giving it a high-tech, cyberpunk vibe. The city street is filled with neon signs, digital billboards, and bustling pedestrians, giving the car a modern and vibrant backdrop. The environment is alive with lights and reflections, with the car reflecting the city’s vibrant colors in its smooth, polished surface. The atmosphere feels both luxurious and high-tech, showcasing the car as the epitome of futuristic automotive design."
image = pipe(
    prompt,
    height=1024,
    width=1024,
    guidance_scale=3.5,
    num_inference_steps=50,
    max_sequence_length=512,
    generator=torch.Generator("cpu").manual_seed(0)
).images[0]
image.save("temple.png")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Open the image you saved
image = Image.open("temple.png")

# Display the image using matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```
